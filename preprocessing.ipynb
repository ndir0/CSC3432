```python
newASF = []
for letter in data.ASF:
    new = str(letter).replace(",", ".")
    newASF.append(new)
data.ASF = newASF
```

```python
data['ASF'] = data['ASF'].astype(float)
```

```python
data.ASF
```

```python
data.iloc[367].ASF
```

```python
num = 0
for i in newASF:
    num += 1
    if i[1] != ".":
        incorrect = str(data.loc[num - 1].ASF)
        while incorrect[1] != ".":
            attempt = float(incorrect)
            attempt /= 10
            incorrect = str(attempt)
        data.at[num - 1, 'ASF'] = float(attempt)
```
